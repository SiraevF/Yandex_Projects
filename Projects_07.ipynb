### Проверка формулы вычисления эффективности обогащения

Эффективность обогащения рассчитывается по формуле:
$$Recovery = \frac{C\times(F-T)}{F\times(C-T)}$$
где:
- $C$ — доля золота в концентрате после флотации/очистки;
- $F$ — доля золота в сырье/концентрате до флотации/очистки;
- $T$ — доля золота в отвальных хвостах после флотации/очистки.

Проверим формулу на наших данных

```python
C = df_train['rougher.output.concentrate_au']
F = df_train['rougher.input.feed_au']
T = df_train['rougher.output.tail_au']
```

```python
recovery_calc = ((C*(F - T))/(F*(C - T)))*100
```

```python
mean_absolute_error(recovery_calc, df_train['rougher.output.recovery'])
```

```python
C_f = df_train['final.output.concentrate_au']
F_f = df_train['rougher.input.feed_au']
T_f = df_train['final.output.tail_au']
```

```python
recovery_calc_f = ((C_f*(F_f - T_f))/(F_f*(C_f - T_f)))*100
```

```python
mean_absolute_error(recovery_calc_f, df_train['final.output.recovery'])
```

MAE имеет очень малое отклонение - это значит, что расчетное значение `rougher.output.recovery`совпадает с данными из столбца.

## Модель

### Функция расчёта sMAPE

```python
def s_mape(target,predict):
    return (2 * abs(target - predict) / (abs(target) + abs(predict))).mean() * 100
```

```python
s_mape_score = make_scorer(s_mape, greater_is_better=False)
```

### Подбор гипперпараметров эффективности обогащения

Проведём подбор гипперпараметров. Используем метод `optuna` для оптимизации поиска. В качестве моделей будут `DecisionTreeRegressor`, `SVR` и `LinearRegression`.

```python
def optuna_optimize(trial):
  
  # задаем пространства поиска гиперпараметров
    if modelType == 'tree':
        min_samples_leaf = trial.suggest_int('min_samples_leaf', 2, 7)
        max_depth = trial.suggest_int('max_depth', 1, 60)

        # создаем модель
        model = DecisionTreeRegressor(max_depth=max_depth,
                                              min_samples_leaf=min_samples_leaf,
                                              random_state=RANDOM_STATE
                                              )
        
    elif modelType == 'ran_for':
        n_estimators = trial.suggest_int('n_estimators', 1, 200)
        min_samples_leaf = trial.suggest_int('min_samples_leaf', 1, 5)
        max_depth = trial.suggest_int('max_depth', 1, 60)

        # создаем модель
        model = RandomForestRegressor(n_estimators=n_estimators,
                                              max_depth=max_depth,
                                              min_samples_leaf=min_samples_leaf,
                                              random_state=RANDOM_STATE,
                                              n_jobs=-1
                                              )
        
    elif modelType == 'svm':
        C = trial.suggest_float('C', 1e-2, 1e2,log=True)
        gamma = trial.suggest_float('gamma', 1e-3, 1e3,log=True)

        # создаем модель
        model = svm.SVR(kernel='rbf', C=C,gamma=gamma)

  # обучаем модель
    if target_name == 'rougher':
        score = cross_val_score(model, features_train, target_train['rougher.output.recovery'], cv = 5, scoring=s_mape_score, n_jobs = -1).mean()  
    elif target_name == 'final':
        score = cross_val_score(model, features_train, target_train['final.output.recovery'], cv = 5, scoring=s_mape_score, n_jobs = -1).mean()  
    return -score
```

```python
%%time
modelType = 'tree'
target_name = 'rougher'
study_tree = optuna.create_study(study_name="Tree",
                                   direction="minimize"
                                   )
# ищем лучшую комбинацию гиперпараметров n_trials раз
study_tree.optimize(optuna_optimize, n_trials=10)

# выводим результаты на обучающей выборке
print("Наилучшие значения гиперпараметров {}".format(study_tree.best_params))
print("s_MAPE: {:.2f}".format(study_tree.best_value))
```

```python
%%time
modelType = 'svm'
target_name = 'rougher'
study_svm = optuna.create_study(study_name="SVM",
                                   direction="minimize"
                                   )
# ищем лучшую комбинацию гиперпараметров n_trials раз
study_svm.optimize(optuna_optimize, n_trials=10)

# выводим результаты на обучающей выборке
print("Наилучшие значения гиперпараметров {}".format(study_svm.best_params))
print("s_MAPE: {:.2f}".format(study_svm.best_value))
```
